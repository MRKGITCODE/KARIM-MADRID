Introduction

This project delves into a comprehensive analysis of a credit card dataset sourced from Kaggle, utilizing the power of data visualization a The aim is to uncover valuable insights into customer behavior and explore the relationships between various gender ,demographic, financial, and product-related factors present in the dataset.
Work to do from the dataset scatterplot : The scatter plot presents an inverse relationship between the credit limit and the average utilization ratio. It utilizes a power regression model
Data Visualization: With Python Matplotlib and seaborn  libarary  robust visualization capabilities will be leveraged to create insightful graphs, charts, and interactive dashboards.
Let's explore various aspects of the credit card dataset through different scatterplot visualizations:

Amount vs. Date Scatterplot:
This scatterplot will show how the transaction amounts vary over time, providing insights into spending trends and patterns.

Amount vs. City Scatterplot:
By plotting the transaction amounts against cities, we can analyze spending behavior across different locations and identify any geographical trends.

Amount vs. Card Type Scatterplot:
This scatterplot will help us understand how transaction amounts vary based on the type of credit card used, providing insights into consumer preferences and card usage patterns.

Amount vs. Gender Scatterplot:
By plotting transaction amounts against gender, we can explore any disparities in spending behavior between male and female cardholders.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Credit.csv')
data.head()

```

PIE PLOT TO SHOW The Highlight Table illustrates a clear preference among credit card holders for the various card types.

```python
# Count the occurrences of each card type
card_type_counts = data['Card Type'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 6))
plt.pie(card_type_counts, labels=card_type_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Card Types')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['Amount'], data['Exp Type'])
plt.title('Scatter Plot of Amount vs Exp Type')
plt.xlabel('Amount')
plt.ylabel('Expense Type')
plt.grid(True)
plt.show()
```

```python
# You may need to parse the 'Date' column into datetime format
data['Date'] = pd.to_datetime(data['Date'])

# Scatterplot of Amount vs. Date
plt.figure(figsize=(10, 6))
plt.scatter(data['Date'], data['Amount'], alpha=0.5)
plt.title('Amount vs. Date')
plt.xlabel('Date')
plt.ylabel('Amount')
plt.show()
```

```python
# Scatterplot of Amount vs. City
plt.figure(figsize=(10, 6))
plt.scatter(data['City'], data['Amount'], alpha=0.5)
plt.title('Amount vs. City')
plt.xlabel('City')
plt.ylabel('Amount')
plt.xticks(rotation=45)
plt.show()
```

```python
# Scatterplot of Amount vs. Card Type
plt.figure(figsize=(10, 6))
plt.scatter(data['Card Type'], data['Amount'], alpha=0.5)
plt.title('Amount vs. Card Type')
plt.xlabel('Card Type')
plt.ylabel('Amount')
plt.show()
```

```python
# Scatterplot of Amount vs. Exp Type
plt.figure(figsize=(10, 6))
plt.scatter(data['Exp Type'], data['Amount'], alpha=0.5)
plt.title('Amount vs. Exp Type')
plt.xlabel('Exp Type')
plt.ylabel('Amount')
plt.show()
```

```python
# Scatterplot of Amount vs. Gender
plt.figure(figsize=(10, 6))
plt.scatter(data['Gender'], data['Amount'], alpha=0.5)
plt.title('Amount vs. Gender')
plt.xlabel('Gender')
plt.ylabel('Amount')
plt.show()
```

```python
# Group by City and calculate the total amount
city_amount = data.groupby('City')['Amount'].sum().reset_index()

# Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='City', y='Amount', hue='City', size='Amount', sizes=(50, 500))
plt.title('Amount vs. City (Grouped)')
plt.xlabel('City')
plt.ylabel('Amount')
plt.xticks(rotation=45)
plt.legend().remove()
# Removing the legend since it's redundant due to grouping
plt.show()
```

```python
data['Date'] = pd.to_datetime(data['Date'])

# Scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Amount', y='Card Type')
plt.title('Scatter Plot of Amount vs. Card Type ')
plt.xlabel('Amount')
plt.ylabel('Card Type')
plt.show()
```

```python
data['Date'] = pd.to_datetime(data['Date'])

# Scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='City', y='Card Type')
plt.title('Scatter Plot of Amount vs. Card Type ')
plt.xlabel('City')
plt.ylabel('Card Type')
plt.show()
```

These plots will give you an idea of how the transaction amounts are distributed with respect to the chosen variables. Remember that scatterplots are useful for visualizing relationships between numerical variables, but they might not be suitable for categorical variables with many unique values. In those cases, you might want to use other types of plots like bar charts or box plots.
